```python
#SI HAY IMAGENES QUE NO TIENEN LABEL, EJECUTAR ESTO
import os #librería para crear carpetas
import cv2
import shutil #librería para copiar archivos
import random
carpetaPrincipal="DB_STEREN_YOLO_UNIDA_1500"
carpetaImagenesYLabels= "DB_YOLO_CLEAN_"+carpetaPrincipal
#carpetaImagenesGenerada="DB_YOLO_"+carpetaPrincipal

archivos=os.listdir(carpetaPrincipal) #extraemos todos los archivos
os.makedirs(carpetaImagenesYLabels) #creamos la carpeta

#Variable para la ruta al directorio
path = carpetaPrincipal
 
#Lista vacia para incluir los ficheros
lstFiles = []

#Lista con todos los ficheros del directorio:
lstDir = os.walk(path)   #os.walk()Lista directorios y ficheros

#Crea una lista de los ficheros txt que existen en el directorio y los incluye a la lista.
 
for root, dirs, files in lstDir:
    for fichero in files:
        (nombreFichero, extension) = os.path.splitext(fichero)
        if(extension == ".txt"):
            lstFiles.append(nombreFichero+extension)
            #print (nombreFichero+extension)
             

lstFiles=random.sample(lstFiles, len(lstFiles))                
print("Solo labels-txt: ",len(lstFiles) )#SOLO JPG       
for i in lstFiles:
    label=i
    #print(imagen)
    nombreImagen=label[:-4]+".jpg"
    shutil.copy(carpetaPrincipal+"/"+label, carpetaImagenesYLabels+'/'+label)
    shutil.copy(carpetaPrincipal+"/"+nombreImagen, carpetaImagenesYLabels+'/'+nombreImagen)

print("terminó!")         
```

```python
#SI TODO ESTÁ EN ORDEN
import os #librería para crear carpetas
import cv2
import shutil #librería para copiar archivos
import random
carpetaPrincipal="imagenes"
carpetaImagenesGenerada="DB_YOLO_"+carpetaPrincipal

archivos=os.listdir(carpetaPrincipal) #extraemos todos los archivos
os.makedirs(carpetaImagenesGenerada) #creamos la carpeta

train=0.7
validation=0.2
test=0.1

countImageNombre=0
countSegundos=0

carpetasDataset=["train","test"]
carpetaTrain=["images","labels"]
carpetasRaices=["train","val"]

for i in carpetasDataset:
    os.makedirs(carpetaImagenesGenerada+"/"+i) #creamos la carpeta
    for j in carpetaTrain:
        os.makedirs(carpetaImagenesGenerada+"/"+i+"/"+j) #creamos la carpeta
        for k in carpetasRaices:
            if(i=="train"):
                os.makedirs(carpetaImagenesGenerada+"/"+i+"/"+j+"/"+k) #creamos la carpeta
            
#Variable para la ruta al directorio
path = carpetaPrincipal

#Lista vacia para incluir los ficheros
lstFiles = []
 
#Lista con todos los ficheros del directorio:
lstDir = os.walk(path)   #os.walk()Lista directorios y ficheros

#Crea una lista de los ficheros jpg que existen en el directorio y los incluye a la lista.
 
for root, dirs, files in lstDir:
    for fichero in files:
        (nombreFichero, extension) = os.path.splitext(fichero)
        if(extension == ".jpg"):
            lstFiles.append(nombreFichero+extension)
            #print (nombreFichero+extension)
             

lstFiles=random.sample(lstFiles, len(lstFiles))                
print("Solo imagenes: ",len(lstFiles) )#SOLO JPG       
cantidadArchivos=len(archivos)/2 #cantidad de archivos en total


cantTrain=int(cantidadArchivos*train)
cantValidation=int(cantidadArchivos*validation)
cantTest=int(cantidadArchivos*test)

print("total imagenes en train: ",cantTrain)
print("total imagenes en validation: ",cantValidation)
print("total imagenes en test: ",cantTest)

countTrain=0
countValidation=0
counTest=0

flag=0
for i in lstFiles:
    imagen=i
    nombreTexto=imagen[:-4]+".txt"
    if(flag==0):
        if(countTrain<cantTrain):
            shutil.move(carpetaPrincipal+"/"+imagen, carpetaImagenesGenerada+"/train/images/train/"+imagen)
            shutil.move(carpetaPrincipal+"/"+nombreTexto, carpetaImagenesGenerada+"/train/labels/train/"+nombreTexto)
            countTrain=countTrain+1
        else:
            flag=1
            
    if(flag==1):
        if(countValidation<cantValidation):
            shutil.move(carpetaPrincipal+"/"+imagen, carpetaImagenesGenerada+"/train/images/val/"+imagen)
            shutil.move(carpetaPrincipal+"/"+nombreTexto, carpetaImagenesGenerada+"/train/labels/val/"+nombreTexto)
            countValidation=countValidation+1
        else:
            flag=2
    if(flag==2):
        if(counTest<cantTest):
            shutil.move(carpetaPrincipal+"/"+imagen, carpetaImagenesGenerada+"/test/images/"+imagen)
            shutil.move(carpetaPrincipal+"/"+nombreTexto, carpetaImagenesGenerada+"/test/labels/"+nombreTexto)
            counTest=counTest+1

print("terminó!")
```
